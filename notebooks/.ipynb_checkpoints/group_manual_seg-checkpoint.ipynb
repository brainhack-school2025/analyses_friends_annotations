```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import plotly.express as px
```

```python
df= pd.read_csv("../output/manual_segmentation_all.tsv", sep='\t')
```

```python
#add frames to onset and offset time
fps = 29.97
df['onset_frame'] = (df['onset']*fps).round().astype(int)
df['offset_frame'] = (df['offset'] * fps).round().astype(int)
df["season"] = df["episode"].str.extract(r's(\d+)', expand=False).astype(int)
```

```python
num_segments = df.groupby("episode_full").size().reset_index(name='num_segments')
```

```python
plt.figure(figsize=(20, 8))
ax = sns.barplot(data=num_segments, x="episode_full", y="num_segments", palette="viridis")
ax.set_xticks(ax.get_xticks()[::5])  # show every 5th label
ax.set_xticklabels(ax.get_xticklabels(), rotation=45, ha="right")
plt.title("Number of segments per episode")
plt.ylabel("n of segments")
plt.xlabel("Episodes")
plt.tight_layout()
plt.show()
```

```python
#average duration of scenes per episode for manual segementation
mean_duration = df.groupby("episode_full")["duration"].mean().reset_index()


plt.figure(figsize=(20, 8))
ax = sns.barplot(data=mean_duration, x="episode_full", y="duration", palette="viridis")
ax.set_xticks(ax.get_xticks()[::5])  # show every 5th label
ax.set_xticklabels(ax.get_xticklabels(), rotation=45, ha="right")
plt.title("Average duration of segments per episode")
plt.ylabel("Average duration (s)")
plt.xlabel("Episodes")
plt.tight_layout()
plt.show()
```

```python
def boolean_True_plotter(list_of_columns, title, x_label, y_label) :
    counts = df[list_of_columns].sum()
    max_y = counts.max() 

    plt.figure(figsize=(20, 10))
    plt.bar(counts.index, counts.values)
    plt.title(title)
    plt.xlabel(x_label)
    plt.ylabel(y_label)
    plt.yticks(np.arange(0, max_y + 50, 50))
    
    plt.show()
    
```

```python
location_columns = ['loc_apt1_Mon_Rach', 'loc_apt2_Chan_Joey', 'loc_apt3_Ross', 'loc_apt4_Pheob_Rach',	'loc_coffeeshop', 'loc_outside', 'loc_other'] 
boolean_True_plotter(location_columns, 'Number of Segments per Location', 'Location', 'Number of segments')
```

```python
onbound_columns = ['ONbond_location', 'ONbond_charact_entry', 'ONbond_charact_leave', 'ONbond_time_jump', 'ONbond_goal_change', 'ONbond_music_transit',	'ONbond_theme_song','ONbond_end']
boolean_True_plotter(onbound_columns, 'Number of segments per ONbound type', 'Onbound type', 'Number of segments')
```

```python
offbound_columns = ['OFFbond_location',	'OFFbond_charact_entry', 'OFFbond_charact_leave',	'OFFbond_time_jump', 'OFFbond_goal_change',	'OFFbond_music_transit',	'OFFbond_theme_song',	'OFFbond_end']
boolean_True_plotter(offbound_columns, 'Number of segments per OFFbound type', 'Offbound type', 'Number of segments')
```

```python
location_cols = [col for col in df.columns if col.startswith('loc_')]
id_vars = ['duration', 'season',] 
df_melted = df.melt(id_vars=['duration', 'season', 'global_segment'],
                    value_vars=location_cols,
                    var_name='location',
                    value_name='is_location')

df_true = df_melted[df_melted['is_location'] == True]



avg_duration = df_true.groupby('location')['duration'].mean().reset_index()


```

```python
plt.figure(figsize=(10,6))
sns.barplot(data=avg_duration, x='location', y='duration', palette='viridis')
plt.xticks(rotation=45, ha='right')
plt.title('Average Duration by Location')
plt.ylabel('Average Duration')
plt.xlabel('Location')
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import plotly.express as px
import ipywidgets as widgets
from IPython.display import display

# Dropdown menu for season selection
season_options = sorted(df_true['season'].unique())
season_dropdown = widgets.Dropdown(
    options=season_options,
    value=season_options[0],
    description="Season:"
)

# Function to generate the violin plot
def update_violin(selected_season):
    # Filter for selected season
    filtered_df = df_true[df_true['season'] == selected_season]

    # Create violin plot
    fig = px.violin(
        filtered_df,
        x="location",            # changed from "episode_short"
        y="duration",            # same: scene/segment duration
        color="location",        # color per location
        hover_data=["global_segment", "duration"],  # changed from scene_number
        box=True,
        points="all",
        title=f"Segment Duration by Location – Season {selected_season}"
    )

    fig.update_layout(
        xaxis_title="Location",
        yaxis_title="Duration (seconds)",
        width=1200,
        height=600,
        showlegend=False
    )

    fig.show()

# Link widget to function
widgets.interact(update_violin, selected_season=season_dropdown)
```

```python
import pandas as pd
import plotly.express as px
import ipywidgets as widgets
from IPython.display import display

# Dropdown menu for season selection
season_options = sorted(df['season'].unique())
season_dropdown = widgets.Dropdown(
    options=season_options,
    value=season_options[0],
    description="Season:"
)

# Function to generate the violin plot
def update_violin(selected_season):
    # Filter for selected season
    filtered_df = df[df['season'] == selected_season]

    fig = px.scatter(
    filtered_df,
    x="global_segment",
    y="duration",
    title=f"Segment Duration over Time – Season {selected_season}"
)



    fig.update_layout(
        xaxis_title="segment",
        yaxis_title="Duration (seconds)",
        width=1200,
        height=600,
        showlegend=False
    )

    fig.show()

# Link widget to function
widgets.interact(update_violin, selected_season=season_dropdown)
```
