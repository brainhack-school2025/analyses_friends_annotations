```python
import numpy as np
import pandas as pd
import os
from glob import glob
base_path = "/mnt/c/Temp/analyses_friends_annotations/friends_annotations/annotation_results/DeepgazeMr"
tsv_files = glob(os.path.join(base_path, '**', '*.tsv'), recursive=True)
all_data = []

for file in tsv_files:

    df = pd.read_csv(file, sep='\t')

    # Extract episode name from filename
    episode = os.path.basename(file).replace('_maxpeak_coord.tsv', '')
    df['episode'] = episode

    all_data.append(df)

# Combine all data into one dataframe for maxpeak
df_max_peak = pd.concat(all_data, ignore_index=True)
#add total frames
df_max_peak['frame'] = df_max_peak.index + 1

# Ensure sorting by episode and global frame (just in case)
df_max_peak = df_max_peak.sort_values(by=["episode", "frame"]).reset_index(drop=True)

# Add episode-relative frame number that resets for each episode
df_max_peak["episode_frame"] = df_max_peak.groupby("episode").cumcount() + 1

#import scenes data tsv




```
